## 1. Model Framing and Problem Definition

### Business Context
The primary business objective is to predict which users are likely to drop off (stop watching) movies before completion. This prediction capability will support several key business decisions:

- **Content Recommendation**: Personalize movie recommendations to reduce dropoff likelihood
- **User Experience Optimization**: Identify users at risk of dropping off for targeted interventions
- **Content Strategy**: Understand which movie characteristics and user behaviors predict completion
- **Resource Allocation**: Focus retention efforts on users most likely to benefit

### Machine Learning Objective
**Primary Goal**: Predict the probability that a user will drop off a movie before completion

**Problem Type**: Binary classification
- **Positive Class (1)**: User tends to drop off movies
- **Negative Class (0)**: User tends to complete movies

### Success Criteria
Given the business context and dataset constraints, success will be measured by:

1. **Primary Metric**: F1-Score > 0.70 (balances precision and recall)
2. **Secondary Metrics**: 
   - Precision > 0.65 (minimize false positives - incorrectly predicting dropoff)
   - Recall > 0.60 (minimize false negatives - missing actual dropoffs)
   - AUC-ROC > 0.75 (overall discriminative ability)

### Model Constraints
- **Interpretability**: High importance for business stakeholder understanding
- **Real-time Requirements**: Model should make predictions within 100ms
- **Feature Availability**: All features must be available at prediction time
- **Data Privacy**: No personally identifiable information in features

## 2. Target Definition and Validation

```python
# Validate the target variable
print("=== TARGET VARIABLE VALIDATION ===")
print(f"Target variable name: 'target_dropoff'")
print(f"Target variable exists: {'target_dropoff' in df.columns}")

if 'target_dropoff' in df.columns:
    y = df['target_dropoff']
    
    # Basic statistics
    print(f"\nTarget Variable Statistics:")
    print(f"  Total samples: {len(y)}")
    print(f"  Non-null samples: {y.notna().sum()}")
    print(f"  Missing values: {y.isna().sum()}")
    print(f"  Data type: {y.dtype}")
    
    # Value distribution
    print(f"\nTarget Distribution:")
    value_counts = y.value_counts().sort_index()
    print(value_counts)
    
    # Class balance
    if len(value_counts) == 2:
        positive_class = value_counts.get(1.0, 0)
        negative_class = value_counts.get(0.0, 0)
        total = positive_class + negative_class
        
        print(f"\nClass Balance:")
        print(f"  Positive class (1 - Dropoff): {positive_class} ({positive_class/total:.1%})")
        print(f"  Negative class (0 - Complete): {negative_class} ({negative_class/total:.1%})")
        
        # Check for severe imbalance
        imbalance_ratio = max(positive_class, negative_class) / min(positive_class, negative_class)
        print(f"  Imbalance ratio: {imbalance_ratio:.2f}:1")
        
        if imbalance_ratio > 3:
            print("  ⚠️  WARNING: Significant class imbalance detected!")
        else:
            print("  ✅ Class balance is acceptable")
    
    # Visualize target distribution
    plt.figure(figsize=(10, 6))
    
    # Bar plot
    plt.subplot(1, 2, 1)
    value_counts.plot(kind='bar', color=['lightcoral', 'lightblue'])
    plt.title('Target Variable Distribution')
    plt.xlabel('Target Value')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    
    # Pie chart
    plt.subplot(1, 2, 2)
    labels = ['Complete Movies (0)', 'Dropout Movies (1)']
    colors = ['lightcoral', 'lightblue']
    plt.pie(value_counts.values, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
    plt.title('Target Distribution')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("❌ Target variable 'target_dropoff' not found in dataset!")
    print("Available columns:", df.columns.tolist())
```

## 3. Feature Selection and Extraction

### Feature Selection Strategy
Based on the feature engineering work, we need to select the optimal subset of features for modeling. The strategy will be:

1. **Exclude non-predictive features**: Remove identifiers, timestamps, and metadata
2. **Select high-importance features**: Use the feature selection results from the engineering phase
3. **Ensure interpretability**: Prioritize features that stakeholders can understand
4. **Avoid multicollinearity**: Remove highly correlated features
5. **Consider computational efficiency**: Balance predictive power with model complexity

```python
# Identify available features
print("=== FEATURE IDENTIFICATION ===")
all_columns = df.columns.tolist()
print(f"Total columns in dataset: {len(all_columns)}")

# Exclude non-predictive columns
exclude_columns = [
    'target_dropoff',  # Target variable
    'Timestamp',       # Timestamp
    # Add any other columns to exclude
]

# Find columns to exclude that exist in the dataset
actual_exclude = [col for col in exclude_columns if col in all_columns]
print(f"Columns to exclude: {actual_exclude}")

# Get potential feature columns
potential_features = [col for col in all_columns if col not in actual_exclude]
print(f"Potential features: {len(potential_features)}")

# Analyze feature types
print(f"\n=== FEATURE ANALYSIS ===")
feature_df = df[potential_features]

# Data types
print("Data types distribution:")
print(feature_df.dtypes.value_counts())

# Missing values
missing_values = feature_df.isnull().sum()
features_with_missing = missing_values[missing_values > 0]
print(f"\nFeatures with missing values: {len(features_with_missing)}")
if len(features_with_missing) > 0:
    print(features_with_missing.head())

# Zero variance features
zero_variance_features = []
for col in feature_df.columns:
    if feature_df[col].nunique() <= 1:
        zero_variance_features.append(col)

print(f"\nZero variance features: {len(zero_variance_features)}")
if zero_variance_features:
    print(zero_variance_features)

# Remove zero variance features
if zero_variance_features:
    potential_features = [col for col in potential_features if col not in zero_variance_features]
    print(f"Features after removing zero variance: {len(potential_features)}")
    feature_df = feature_df.drop(columns=zero_variance_features)
```

```python
# Feature Selection Implementation
print("=== FEATURE SELECTION IMPLEMENTATION ===")

# For this implementation, we'll use a combination of approaches:
# 1. Remove features with high missing values
# 2. Remove highly correlated features
# 3. Select features with reasonable predictive power

# 1. Handle missing values
missing_threshold = 0.1  # Remove features with >10% missing values
high_missing_features = []
for col in feature_df.columns:
    missing_pct = feature_df[col].isnull().mean()
    if missing_pct > missing_threshold:
        high_missing_features.append(col)

if high_missing_features:
    print(f"Removing {len(high_missing_features)} features with >{missing_threshold*100}% missing values")
    feature_df = feature_df.drop(columns=high_missing_features)
    potential_features = [col for col in potential_features if col not in high_missing_features]

# 2. Remove highly correlated features
print(f"\nChecking for highly correlated features...")
numeric_features = feature_df.select_dtypes(include=[np.number]).columns
correlation_threshold = 0.95

if len(numeric_features) > 1:
    corr_matrix = feature_df[numeric_features].corr().abs()
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    highly_correlated_features = []
    for col in upper_triangle.columns:
        if any(upper_triangle[col] > correlation_threshold):
            highly_correlated_features.append(col)
    
    if highly_correlated_features:
        print(f"Removing {len(highly_correlated_features)} highly correlated features")
        feature_df = feature_df.drop(columns=highly_correlated_features)
        potential_features = [col for col in potential_features if col not in highly_correlated_features]

# 3. Basic feature importance using correlation with target
print(f"\nCalculating feature importance...")
feature_importance = []
for col in feature_df.columns:
    if col in feature_df.select_dtypes(include=[np.number]).columns:
        corr = feature_df[col].corr(df['target_dropoff'])
        if not np.isnan(corr):
            feature_importance.append((col, abs(corr)))

# Sort by importance
feature_importance.sort(key=lambda x: x[1], reverse=True)

print(f"Top 10 most important features by correlation:")
for i, (feature, importance) in enumerate(feature_importance[:10]):
    print(f"{i+1:2d}. {feature}: {importance:.4f}")

# Select top features (adjust number as needed)
top_n_features = 50  # Adjust this based on your needs
selected_feature_names = [f[0] for f in feature_importance[:top_n_features]]

print(f"\nSelected {len(selected_feature_names)} features for modeling")
print(f"Feature selection complete!")

# Create final feature matrix
X_final = feature_df[selected_feature_names].copy()
print(f"\nFinal feature matrix shape: {X_final.shape}")

# Handle any remaining missing values
if X_final.isnull().sum().sum() > 0:
    print("Handling remaining missing values...")
    # Fill numeric columns with median
    for col in X_final.select_dtypes(include=[np.number]).columns:
        if X_final[col].isnull().sum() > 0:
            X_final[col].fillna(X_final[col].median(), inplace=True)
    
    # Fill non-numeric columns with mode
    for col in X_final.select_dtypes(exclude=[np.number]).columns:
        if X_final[col].isnull().sum() > 0:
            X_final[col].fillna(X_final[col].mode()[0], inplace=True)

print(f"Final missing values: {X_final.isnull().sum().sum()}")
print(f"Final feature matrix shape: {X_final.shape}")
print(f"Final target variable shape: {y_final.shape}")
```

## 4. Final Dataset Assembly

### Creating the Final Modeling Dataset
Now we combine the selected features with the target variable to create the final dataset ready for machine learning model training.

```python
# Assemble the final modeling dataset
print("=== FINAL DATASET ASSEMBLY ===")

# Create the final modeling dataset
# Combine selected features with target variable
feature_cols = selected_feature_names
target_col = 'target_dropoff'

# Create the final dataset
ml_dataset = pd.DataFrame()
ml_dataset[feature_cols] = X_final
ml_dataset[target_col] = y_final

print(f"Final modeling dataset created!")
print(f"Dataset shape: {ml_dataset.shape}")
print(f"Features: {len(feature_cols)}")
print(f"Target column: {target_col}")

# Final data quality checks
print(f"\n=== FINAL DATA QUALITY CHECKS ===")
print(f"Total samples: {len(ml_dataset)}")
print(f"Total features: {len(feature_cols)}")
print(f"Missing values: {ml_dataset.isnull().sum().sum()}")
print(f"Duplicate rows: {ml_dataset.duplicated().sum()}")
print(f"Data types: {ml_dataset.dtypes.value_counts().to_dict()}")

# Target distribution in final dataset
print(f"\nTarget distribution in final dataset:")
print(ml_dataset[target_col].value_counts().sort_index())

# Feature summary
print(f"\nFeature summary:")
print(f"- Numeric features: {len(ml_dataset.select_dtypes(include=[np.number]).columns) - 1}")  # -1 for target
print(f"- Non-numeric features: {len(ml_dataset.select_dtypes(exclude=[np.number]).columns)}")

# Save the final dataset
output_file = 'ml_dataset_final.csv'
ml_dataset.to_csv(output_file, index=False)
print(f"\n✅ Final modeling dataset saved as: {output_file}")

# Also save X and y separately for convenience
X_final.to_csv('X_final.csv', index=False)
y_final.to_csv('y_final.csv', index=False)
print(f"✅ Feature matrix saved as: X_final.csv")
print(f"✅ Target variable saved as: y_final.csv")

print(f"\n=== DATASET ASSEMBLY COMPLETE ===")
print(f"Ready for model training!")
```

```python
# Create Data Dictionary and Documentation
print("=== DATA DICTIONARY AND DOCUMENTATION ===")

# Feature categories for documentation
feature_categories = {
    'binary_features': [],
    'engineered_features': [],
    'discretized_features': [],
    'encoded_features': [],
    'other_features': []
}

# Categorize features based on naming patterns
for col in feature_cols:
    if col.startswith('which_genres_') or col.startswith('how_do_you_') or col.startswith('where_do_you_') or col.startswith('why_do_you_') or col.startswith('in_general_') or col.startswith('do_you_usually_'):
        feature_categories['binary_features'].append(col)
    elif 'total_' in col or 'ratio' in col or 'interaction' in col:
        feature_categories['engineered_features'].append(col)
    elif 'bins' in col or 'binned' in col:
        feature_categories['discretized_features'].append(col)
    elif '_encoded' in col:
        feature_categories['encoded_features'].append(col)
    else:
        feature_categories['other_features'].append(col)

# Print feature categories
print(f"Feature Categories:")
for category, features in feature_categories.items():
    print(f"- {category.replace('_', ' ').title()}: {len(features)}")

# Create detailed data dictionary
data_dictionary = {
    'feature_name': [],
    'feature_type': [],
    'description': [],
    'data_type': [],
    'unique_values': [],
    'missing_values': []
}

for col in feature_cols:
    data_dictionary['feature_name'].append(col)
    data_dictionary['data_type'].append(str(X_final[col].dtype))
    data_dictionary['unique_values'].append(X_final[col].nunique())
    data_dictionary['missing_values'].append(X_final[col].isnull().sum())
    
    # Determine feature type
    if col in feature_categories['binary_features']:
        data_dictionary['feature_type'].append('Binary')
        data_dictionary['description'].append('Binary survey response (0/1)')
    elif col in feature_categories['engineered_features']:
        data_dictionary['feature_type'].append('Engineered')
        data_dictionary['description'].append('Engineered feature from original data')
    elif col in feature_categories['discretized_features']:
        data_dictionary['feature_type'].append('Discretized')
        data_dictionary['description'].append('Discretized/binned continuous variable')
    elif col in feature_categories['encoded_features']:
        data_dictionary['feature_type'].append('Encoded')
        data_dictionary['description'].append('Label-encoded categorical variable')
    else:
        data_dictionary['feature_type'].append('Other')
        data_dictionary['description'].append('Other feature type')

# Create data dictionary DataFrame
data_dict_df = pd.DataFrame(data_dictionary)

# Save data dictionary
data_dict_df.to_csv('data_dictionary.csv', index=False)
print(f"\n✅ Data dictionary saved as: data_dictionary.csv")

# Display first few rows
print(f"\nData Dictionary Preview:")
print(data_dict_df.head(10))
```
