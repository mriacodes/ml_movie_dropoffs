# Behavioral Movie Seeding Script - Domain-Appropriate Heuristics

This notebook implements **sophisticated behavioral modeling** to augment survey responses with realistic movie-viewing patterns. The goal is to simulate authentic human choices that reflect genuine movie-watching behavior based on demographic, preference, and behavioral signals.

## ðŸŽ¯ Objective
- **Behavioral Realism**: Simulate authentic human movie choices based on survey responses
- **Demographic Alignment**: Match movies to age, education, and viewing frequency patterns
- **Preference Consistency**: Align movie selections with stated genre preferences
- **Behavioral Logic**: Reflect realistic completion vs. abandonment patterns
- **ML-Ready Data**: Create synthetic data that preserves authentic human behavior patterns

## ðŸ§© Behavioral Modeling Approach
- **Finished Movies**: Popular, well-rated, shorter films matching user preferences
- **Unfinished Movies**: Longer, niche, or genre-mismatched films reflecting realistic abandonment patterns
- **Demographic Heuristics**: Age-appropriate content, education-complexity matching
- **Preference Weighting**: Strong bias toward stated genre preferences
- **Contextual Logic**: Viewing frequency influences content sophistication

## ðŸ“Š Data Sources
- **Survey Dataset**: 77 respondent records with demographics, preferences, and dropout behavior
- **Movie Metadata**: 5,000+ movies with ratings, genres, runtime, and popularity metrics

```python
# Load the datasets
survey_df = pd.read_csv('dropoffs_preprocessed.csv')
movies_df = pd.read_csv('cleaned_imdb_data.csv')

print("=== SURVEY DATA ===")
print(f"Survey dataset shape: {survey_df.shape}")
print(f"Columns: {survey_df.columns.tolist()}")

print("\n=== MOVIE DATA ===")
print(f"Movie dataset shape: {movies_df.shape}")
print(f"Columns: {movies_df.columns.tolist()}")

# Check the key column for drop-off behavior
dropoff_column = 'Have you ever started watching a movie but did not finish it?'
print(f"\n=== DROP-OFF RESPONSES ===")
print(survey_df[dropoff_column].value_counts())
print(f"Response distribution:")
print(survey_df[dropoff_column].value_counts(normalize=True))
```

```python
def extract_respondent_preferences(respondent):
    """Extract behavioral preferences from survey respondent data with improved genre matching"""
    preferences = {}
    
    # IMPROVED GENRE PREFERENCE EXTRACTION - Using actual binary column names
    genre_columns = {
        'which_genres_do_you_enjoy_watching_the_most_action': 'Action',
        'which_genres_do_you_enjoy_watching_the_most_comedy': 'Comedy',
        'which_genres_do_you_enjoy_watching_the_most_drama': 'Drama',
        'which_genres_do_you_enjoy_watching_the_most_horror': 'Horror',
        'which_genres_do_you_enjoy_watching_the_most_romance': 'Romance',
        'which_genres_do_you_enjoy_watching_the_most_science_fiction_sci_fi': 'Sci-Fi',
        'which_genres_do_you_enjoy_watching_the_most_documentary': 'Documentary',
        'which_genres_do_you_enjoy_watching_the_most_thriller': 'Thriller',
        'which_genres_do_you_enjoy_watching_the_most_historical': 'History'
    }
    
    preferred_genres = []
    for col, genre in genre_columns.items():
        if respondent.get(col, 0) == 1:  # Binary 1 = preferred
            preferred_genres.append(genre)
    
    # Map some genre names to movie database format
    genre_mapping = {
        'Sci-Fi': 'Science Fiction',
        'History': 'Historical'
    }
    
    # Apply mapping
    mapped_genres = []
    for genre in preferred_genres:
        mapped_genres.append(genre_mapping.get(genre, genre))
    
    preferences['preferred_genres'] = mapped_genres
    
    # EXTRACT GENRES TO AVOID - Using stop watching patterns
    avoid_genres = []
    stop_genre_columns = {
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_action': 'Action',
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_comedy': 'Comedy',
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_drama': 'Drama',
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_horror': 'Horror',
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_romance': 'Romance',
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_science_fiction_sci_fi': 'Sci-Fi',
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_documentary': 'Documentary',
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_thriller': 'Thriller',
        'which_genres_do_you_find_yourself_stopping_more_often_before_finishing_historical': 'History'
    }
    
    for col, genre in stop_genre_columns.items():
        if respondent.get(col, 0) == 1:  # Binary 1 = stops watching this genre
            avoid_genres.append(genre_mapping.get(genre, genre))
    
    preferences['avoid_genres'] = avoid_genres
    
    # Age group behavioral patterns
    age_group = respondent.get('What is your age group?', '')
    age_group_str = str(age_group) if pd.notna(age_group) else ''
    
    if '18-24' in age_group_str:
        preferences['age_preference'] = 'young'
        preferences['complexity_tolerance'] = 'low'
        preferences['mainstream_bias'] = 0.8
        preferences['sophistication_level'] = 'low'
        preferences['arthouse_tolerance'] = 0.1
    elif '25-34' in age_group_str:
        preferences['age_preference'] = 'adult'
        preferences['complexity_tolerance'] = 'medium'
        preferences['mainstream_bias'] = 0.6
        preferences['sophistication_level'] = 'medium'
        preferences['arthouse_tolerance'] = 0.2
    elif '35-44' in age_group_str:
        preferences['age_preference'] = 'adult'
        preferences['complexity_tolerance'] = 'medium'
        preferences['mainstream_bias'] = 0.5
        preferences['sophistication_level'] = 'medium'
        preferences['arthouse_tolerance'] = 0.3
    elif '45-54' in age_group_str:
        preferences['age_preference'] = 'mature'
        preferences['complexity_tolerance'] = 'high'
        preferences['mainstream_bias'] = 0.4
        preferences['sophistication_level'] = 'high'
        preferences['arthouse_tolerance'] = 0.4
    elif '55-64' in age_group_str:
        preferences['age_preference'] = 'mature'
        preferences['complexity_tolerance'] = 'high'
        preferences['mainstream_bias'] = 0.3
        preferences['sophistication_level'] = 'high'
        preferences['arthouse_tolerance'] = 0.5
    elif '65+' in age_group_str:
        preferences['age_preference'] = 'mature'
        preferences['complexity_tolerance'] = 'high'
        preferences['mainstream_bias'] = 0.2
        preferences['sophistication_level'] = 'high'
        preferences['arthouse_tolerance'] = 0.6
    else:
        # Default for unknown age
        preferences['age_preference'] = 'adult'
        preferences['complexity_tolerance'] = 'medium'
        preferences['mainstream_bias'] = 0.5
        preferences['sophistication_level'] = 'medium'
        preferences['arthouse_tolerance'] = 0.3
    
    # Education level influences sophistication
    education = respondent.get("What is the highest level of education you've completed? ", '')
    education_str = str(education) if pd.notna(education) else ''
    
    if 'High school' in education_str:
        preferences['sophistication_level'] = 'low'
        preferences['arthouse_tolerance'] = max(0.1, preferences['arthouse_tolerance'] - 0.1)
        preferences['mainstream_bias'] = min(0.9, preferences['mainstream_bias'] + 0.1)
    elif 'Bachelor' in education_str:
        preferences['sophistication_level'] = 'medium'
    elif 'Graduate' in education_str:
        preferences['sophistication_level'] = 'high'
        preferences['arthouse_tolerance'] = min(0.8, preferences['arthouse_tolerance'] + 0.2)
        preferences['mainstream_bias'] = max(0.2, preferences['mainstream_bias'] - 0.2)
    
    # Viewing frequency affects diversity seeking
    frequency = respondent.get('How often do you watch movies?', '')
    frequency_str = str(frequency) if pd.notna(frequency) else ''
    
    if 'Daily' in frequency_str:
        preferences['diversity_seeking'] = 'high'
        preferences['niche_tolerance'] = 0.6
    elif 'Several times a week' in frequency_str:
        preferences['diversity_seeking'] = 'medium'
        preferences['niche_tolerance'] = 0.4
    elif 'Once a week' in frequency_str:
        preferences['diversity_seeking'] = 'medium'
        preferences['niche_tolerance'] = 0.3
    elif 'A few times a month' in frequency_str:
        preferences['diversity_seeking'] = 'low'
        preferences['niche_tolerance'] = 0.2
    elif 'Rarely' in frequency_str:
        preferences['diversity_seeking'] = 'low'
        preferences['niche_tolerance'] = 0.1
    else:
        preferences['diversity_seeking'] = 'low'
        preferences['niche_tolerance'] = 0.1
    
    return preferences

# Test improved preference extraction
print("=== IMPROVED PREFERENCE EXTRACTION TEST ===")
test_respondents = [0, 1, 2, 5, 10]

for idx in test_respondents:
    if idx < len(survey_df):
        respondent = survey_df.iloc[idx]
        prefs = extract_respondent_preferences(respondent)
        print(f"\nRespondent {idx + 1}:")
        print(f"  Age: {respondent.get('What is your age group?', 'N/A')}")
        
        # Fixed the problematic line
        education_col = "What is the highest level of education you've completed? "
        print(f"  Education: {respondent.get(education_col, 'N/A')}")
        
        print(f"  Frequency: {respondent.get('How often do you watch movies?', 'N/A')}")
        print(f"  Preferred genres: {prefs['preferred_genres']}")
        print(f"  Avoid genres: {prefs['avoid_genres']}")
        print(f"  Age preference: {prefs['age_preference']}")
        print(f"  Sophistication: {prefs['sophistication_level']}")

print("\n=== PREFERENCE EXTRACTION IMPROVEMENT COMPLETE ===")
```

```python
def select_behaviorally_appropriate_movies(movie_pool, respondent, n_movies=5, movie_type='finished', used_movies=None):
    """Improved movie selection with better genre matching and age-appropriate filtering"""
    if used_movies is None:
        used_movies = set()
    
    # Get respondent preferences
    preferences = extract_respondent_preferences(respondent)
    
    # Start with available movies (exclude already used)
    available_movies = movie_pool[~movie_pool['movie_title'].isin(used_movies)].copy()
    
    if len(available_movies) == 0:
        return movie_pool.sample(n=min(n_movies, len(movie_pool)))
    
    # AGE-APPROPRIATE FILTERING
    age_preference = preferences.get('age_preference', 'adult')
    frequency = respondent.get('How often do you watch movies?', '')
    frequency_str = str(frequency) if pd.notna(frequency) else ''
    
    # Filter for young, infrequent viewers
    if age_preference == 'young' and 'Rarely' in frequency_str:
        # Filter out very long movies (>130 min) for young, infrequent viewers
        available_movies = available_movies[available_movies['duration'] <= 130]
        
        # Filter out very low-rated movies (<6.0) for casual viewers
        available_movies = available_movies[available_movies['imdb_score'] >= 6.0]
    
    # Initialize behavioral score
    available_movies['behavioral_score'] = 0.0
    
    # 1. GENRE PREFERENCE SCORING (50% weight) - IMPROVED
    preferred_genres = preferences.get('preferred_genres', [])
    avoid_genres = preferences.get('avoid_genres', [])
    
    if preferred_genres:
        for idx, movie in available_movies.iterrows():
            movie_genres = [g.strip() for g in str(movie['genres']).split(',')]
            
            # Strong bonus for exact genre matches
            exact_matches = sum(1 for pref in preferred_genres 
                              for g in movie_genres 
                              if pref.lower() in g.lower())
            
            if exact_matches > 0:
                available_movies.loc[idx, 'behavioral_score'] += exact_matches * 25  # Up to 50 points
    
    # Penalty for avoid genres
    if avoid_genres:
        for idx, movie in available_movies.iterrows():
            movie_genres = [g.strip() for g in str(movie['genres']).split(',')]
            
            # Penalty for genres the respondent tends to stop watching
            avoid_matches = sum(1 for avoid in avoid_genres 
                               for g in movie_genres 
                               if avoid.lower() in g.lower())
            
            if avoid_matches > 0:
                available_movies.loc[idx, 'behavioral_score'] -= avoid_matches * 15  # Penalty
    
    # 2. AGE APPROPRIATENESS SCORING (20% weight) - Use imdb_score as proxy for mainstream appeal
    score_median = available_movies['imdb_score'].median()
    mainstream_scores = (available_movies['imdb_score'] >= score_median).astype(float)
    
    if age_preference == 'young':
        available_movies['behavioral_score'] += mainstream_scores * 20
    elif age_preference == 'adult':
        available_movies['behavioral_score'] += mainstream_scores * 15
    else:  # mature
        available_movies['behavioral_score'] += mainstream_scores * 10
    
    # 3. SOPHISTICATION SCORING (15% weight)
    sophistication_level = preferences.get('sophistication_level', 'medium')
    rating_75th = available_movies['imdb_score'].quantile(0.75)
    sophistication_scores = (available_movies['imdb_score'] >= rating_75th).astype(float)
    
    if sophistication_level == 'low':
        available_movies['behavioral_score'] += (1 - sophistication_scores) * 15
    elif sophistication_level == 'medium':
        available_movies['behavioral_score'] += 7.5  # neutral
    else:  # high
        available_movies['behavioral_score'] += sophistication_scores * 15
    
    # 4. VIEWING FREQUENCY SCORING (10% weight)
    diversity_seeking = preferences.get('diversity_seeking', 'low')
    genre_diversity = available_movies['genre_count']
    diversity_scores = (genre_diversity >= genre_diversity.median()).astype(float)
    
    if diversity_seeking == 'high':
        available_movies['behavioral_score'] += diversity_scores * 10
    elif diversity_seeking == 'medium':
        available_movies['behavioral_score'] += 5  # neutral
    # low diversity seeking gets no bonus
    
    # 5. RUNTIME APPROPRIATENESS SCORING (5% weight)
    runtime_median = available_movies['duration'].median()
    
    if movie_type == 'finished':
        # Shorter movies get higher scores for finished category
        runtime_scores = (available_movies['duration'] <= runtime_median).astype(float)
        available_movies['behavioral_score'] += runtime_scores * 5
    else:  # unfinished
        # Longer movies get higher scores for unfinished category
        runtime_scores = (available_movies['duration'] > runtime_median).astype(float)
        available_movies['behavioral_score'] += runtime_scores * 5
    
    # Add small random component to prevent identical selections
    np.random.seed(None)  # Use current time as seed
    available_movies['behavioral_score'] += np.random.uniform(0, 3, len(available_movies))
    
    # Select top movies based on behavioral score
    if len(available_movies) < n_movies:
        return available_movies
    
    selected_movies = available_movies.nlargest(n_movies, 'behavioral_score')
    
    return selected_movies

print("=== MOVIE CATEGORIZATION FOR IMPROVED SEEDING ===")

# Categorize movies for behavioral selection using correct column names
duration_threshold = 130  # minutes
rating_threshold = 6.0

print(f"Duration threshold: {duration_threshold} minutes")
print(f"Rating threshold: {rating_threshold}")

# Movies more likely to be FINISHED (mainstream, shorter, higher-rated)
finished_movies = movies_df[
    (movies_df['duration'] <= duration_threshold) &
    (movies_df['imdb_score'] >= rating_threshold)
].copy()

# Movies more likely to be UNFINISHED (longer, lower-rated, challenging)
unfinished_movies = movies_df[
    (movies_df['duration'] > duration_threshold) |
    (movies_df['imdb_score'] < rating_threshold)
].copy()

print(f"\nFinished movies pool: {len(finished_movies)} movies")
print(f"Unfinished movies pool: {len(unfinished_movies)} movies")

print("\n=== TESTING IMPROVED MOVIE SELECTION ===")

# Test the improved selection function
for i in range(3):
    print(f"\n--- Testing Respondent {i+1} ---")
    respondent = survey_df.iloc[i]
    preferences = extract_respondent_preferences(respondent)
    
    print(f"Age: {respondent.get('What is your age group?', 'N/A')}")
    
    # Fixed the problematic line
    education_col = "What is the highest level of education you've completed? "
    print(f"Education: {respondent.get(education_col, 'N/A')}")
    
    print(f"Frequency: {respondent.get('How often do you watch movies?', 'N/A')}")
    print(f"Preferred genres: {preferences['preferred_genres']}")
    print(f"Avoid genres: {preferences['avoid_genres']}")
    
    # Test finished movies
    finished_selection = select_behaviorally_appropriate_movies(
        finished_movies, respondent, n_movies=5, movie_type='finished'
    )
    print(f"Finished movies: {finished_selection['movie_title'].tolist()}")
    
    # Test unfinished movies
    unfinished_selection = select_behaviorally_appropriate_movies(
        unfinished_movies, respondent, n_movies=5, movie_type='unfinished'
    )
    print(f"Unfinished movies: {unfinished_selection['movie_title'].tolist()}")

print("\n=== IMPROVED MOVIE SELECTION READY ===")
```

```python
def seed_survey_with_improved_behavioral_realism(survey_df, finished_movies, unfinished_movies):
    """
    Improved behavioral seeding with logical consistency validation
    """
    # Create a copy to avoid modifying original
    seeded_df = survey_df.copy()
    
    # Initialize movie columns
    for i in range(1, 6):
        seeded_df[f'finished_{i}'] = None
        seeded_df[f'unfinished_{i}'] = None
    
    # Track global movie usage for diversity
    global_used_movies = set()
    
    print("=== IMPROVED BEHAVIORAL MOVIE SEEDING ===")
    print("Implementing enhanced behavioral modeling with logical consistency")
    
    for idx, respondent in seeded_df.iterrows():
        print(f"Processing respondent {idx + 1}/{len(seeded_df)} with improved behavioral modeling")
        
        # Extract behavioral preferences
        preferences = extract_respondent_preferences(respondent)
        
        # Track movies used by this respondent
        used_movies_this_respondent = set()
        
        # CHECK FOR LOGICAL CONSISTENCY
        dropoff_response = respondent.get('Have you ever started watching a movie but did not finish it?', '')
        dropoff_response_str = str(dropoff_response) if pd.notna(dropoff_response) else ''
        
        if 'No' in dropoff_response_str:
            # For "No" respondents, only assign finished movies
            print(f"  Respondent {idx + 1}: 'No' drop-off - assigning only finished movies")
            
            # SELECT FINISHED MOVIES ONLY
            finished_selection = select_behaviorally_appropriate_movies(
                finished_movies, 
                respondent, 
                n_movies=5, 
                movie_type='finished',
                used_movies=used_movies_this_respondent
            )
            
            # Assign finished movies
            for i, (_, movie) in enumerate(finished_selection.iterrows()):
                if i < 5:  # Ensure we don't exceed 5 movies
                    seeded_df.loc[idx, f'finished_{i+1}'] = movie['movie_title']
                    used_movies_this_respondent.add(movie['movie_title'])
                    global_used_movies.add(movie['movie_title'])
            
            # Leave unfinished columns empty for "No" respondents
            for i in range(1, 6):
                seeded_df.loc[idx, f'unfinished_{i}'] = None
                
        else:
            # For "Yes" respondents, assign both finished and unfinished movies
            print(f"  Respondent {idx + 1}: 'Yes' drop-off - assigning both finished and unfinished movies")
            
            # SELECT FINISHED MOVIES with behavioral appropriateness
            finished_selection = select_behaviorally_appropriate_movies(
                finished_movies, 
                respondent, 
                n_movies=5, 
                movie_type='finished',
                used_movies=used_movies_this_respondent
            )
            
            # Assign finished movies
            for i, (_, movie) in enumerate(finished_selection.iterrows()):
                if i < 5:  # Ensure we don't exceed 5 movies
                    seeded_df.loc[idx, f'finished_{i+1}'] = movie['movie_title']
                    used_movies_this_respondent.add(movie['movie_title'])
                    global_used_movies.add(movie['movie_title'])
            
            # SELECT UNFINISHED MOVIES with behavioral appropriateness
            unfinished_selection = select_behaviorally_appropriate_movies(
                unfinished_movies, 
                respondent, 
                n_movies=5, 
                movie_type='unfinished',
                used_movies=used_movies_this_respondent
            )
            
            # Assign unfinished movies
            for i, (_, movie) in enumerate(unfinished_selection.iterrows()):
                if i < 5:  # Ensure we don't exceed 5 movies
                    seeded_df.loc[idx, f'unfinished_{i+1}'] = movie['movie_title']
                    used_movies_this_respondent.add(movie['movie_title'])
                    global_used_movies.add(movie['movie_title'])
    
    print(f"Total unique movies used: {len(global_used_movies)}")
    print("=== IMPROVED BEHAVIORAL SEEDING COMPLETE ===")
    
    return seeded_df

# Execute improved behavioral seeding
print("Starting IMPROVED BEHAVIORAL movie seeding with enhanced logic...")
seeded_survey = seed_survey_with_improved_behavioral_realism(survey_df, finished_movies, unfinished_movies)

print(f"\nOriginal survey shape: {survey_df.shape}")
print(f"Seeded survey shape: {seeded_survey.shape}")

# Quick validation
dropoff_column = 'Have you ever started watching a movie but did not finish it?'
no_respondents = seeded_survey[seeded_survey[dropoff_column] == 'No']
print(f"\nValidation: {len(no_respondents)} respondents said 'No' to drop-off")

# Check if "No" respondents have empty unfinished columns
for idx in no_respondents.index:
    unfinished_count = sum(1 for i in range(1, 6) if pd.notna(seeded_survey.loc[idx, f'unfinished_{i}']))
    if unfinished_count > 0:
        print(f"WARNING: Respondent {idx + 1} said 'No' but has {unfinished_count} unfinished movies")
    else:
        print(f"âœ“ Respondent {idx + 1} logical consistency maintained")

print("\n=== IMPROVED SEEDING VALIDATION COMPLETE ===")
```

```python
# COMPREHENSIVE VALIDATION OF IMPROVED SEEDING
print("=== COMPREHENSIVE SEEDING QUALITY VALIDATION ===")

def validate_seeding_quality(seeded_df):
    """Comprehensive validation of seeding quality"""
    validation_results = {}
    
    # 1. LOGICAL CONSISTENCY CHECK
    print("\n1. LOGICAL CONSISTENCY VALIDATION")
    dropoff_column = 'Have you ever started watching a movie but did not finish it?'
    
    no_respondents = seeded_df[seeded_df[dropoff_column] == 'No']
    logical_errors = 0
    
    for idx in no_respondents.index:
        unfinished_count = sum(1 for i in range(1, 6) if pd.notna(seeded_df.loc[idx, f'unfinished_{i}']))
        if unfinished_count > 0:
            logical_errors += 1
            print(f"  ERROR: Respondent {idx + 1} said 'No' but has {unfinished_count} unfinished movies")
    
    if logical_errors == 0:
        print("  âœ“ All 'No' respondents have empty unfinished columns")
    
    validation_results['logical_errors'] = logical_errors
    
    # 2. GENRE PREFERENCE ALIGNMENT CHECK
    print("\n2. GENRE PREFERENCE ALIGNMENT VALIDATION")
    
    genre_alignment_scores = []
    
    for idx in range(min(10, len(seeded_df))):  # Check first 10 respondents
        respondent = seeded_df.iloc[idx]
        preferences = extract_respondent_preferences(respondent)
        preferred_genres = preferences.get('preferred_genres', [])
        
        if preferred_genres:
            # Check finished movies for genre alignment
            finished_movies_assigned = []
            for i in range(1, 6):
                movie = respondent[f'finished_{i}']
                if pd.notna(movie):
                    finished_movies_assigned.append(movie)
            
            # Calculate alignment score
            aligned_movies = 0
            for movie_title in finished_movies_assigned:
                movie_data = movies_df[movies_df['movie_title'] == movie_title]
                if not movie_data.empty:
                    movie_genres = [g.strip() for g in str(movie_data.iloc[0]['genres']).split(',')]
                    if any(pref.lower() in g.lower() for pref in preferred_genres for g in movie_genres):
                        aligned_movies += 1
            
            if len(finished_movies_assigned) > 0:
                alignment_score = aligned_movies / len(finished_movies_assigned)
                genre_alignment_scores.append(alignment_score)
                
                print(f"  Respondent {idx + 1}: {aligned_movies}/{len(finished_movies_assigned)} movies match preferences ({alignment_score:.1%})")
                print(f"    Preferred: {preferred_genres}")
                print(f"    Finished: {finished_movies_assigned}")
    
    avg_alignment = np.mean(genre_alignment_scores) if genre_alignment_scores else 0
    print(f"  Average genre alignment: {avg_alignment:.1%}")
    validation_results['genre_alignment'] = avg_alignment
    
    # 3. AGE-APPROPRIATENESS CHECK
    print("\n3. AGE-APPROPRIATENESS VALIDATION")
    
    age_appropriateness_issues = 0
    
    for idx in range(min(10, len(seeded_df))):
        respondent = seeded_df.iloc[idx]
        age_group = respondent.get('What is your age group?', '')
        frequency = respondent.get('How often do you watch movies?', '')
        
        if '18-24' in str(age_group) and 'Rarely' in str(frequency):
            # Check for long/complex movies
            finished_movies_assigned = []
            for i in range(1, 6):
                movie = respondent[f'finished_{i}']
                if pd.notna(movie):
                    finished_movies_assigned.append(movie)
            
            for movie_title in finished_movies_assigned:
                movie_data = movies_df[movies_df['movie_title'] == movie_title]
                if not movie_data.empty:
                    duration = movie_data.iloc[0]['duration']
                    rating = movie_data.iloc[0]['imdb_score']
                    
                    if duration > 140 or rating < 5.5:  # Very long or low-rated movies
                        age_appropriateness_issues += 1
                        print(f"  WARNING: Respondent {idx + 1} (18-24, Rarely) assigned '{movie_title}' ({duration}min, {rating}/10)")
    
    if age_appropriateness_issues == 0:
        print("  âœ“ Age-appropriate content assignment maintained")
    
    validation_results['age_appropriateness_issues'] = age_appropriateness_issues
    
    # 4. MOVIE REUSE RATE CALCULATION
    print("\n4. MOVIE REUSE RATE ANALYSIS")
    
    all_finished_movies = []
    all_unfinished_movies = []
    
    for i in range(1, 6):
        finished_movies = seeded_df[f'finished_{i}'].dropna().tolist()
        unfinished_movies = seeded_df[f'unfinished_{i}'].dropna().tolist()
        
        all_finished_movies.extend(finished_movies)
        all_unfinished_movies.extend(unfinished_movies)
    
    all_assigned_movies = all_finished_movies + all_unfinished_movies
    
    total_assignments = len(all_assigned_movies)
    unique_movies = len(set(all_assigned_movies))
    reuse_rate = ((total_assignments - unique_movies) / total_assignments) * 100
    
    print(f"  Total movie assignments: {total_assignments}")
    print(f"  Unique movies assigned: {unique_movies}")
    print(f"  Movie reuse rate: {reuse_rate:.1f}%")
    
    validation_results['reuse_rate'] = reuse_rate
    validation_results['total_assignments'] = total_assignments
    validation_results['unique_movies'] = unique_movies
    
    return validation_results

# Run comprehensive validation
validation_results = validate_seeding_quality(seeded_survey)

print("\n=== OVERALL VALIDATION SUMMARY ===")
print(f"âœ“ Logical consistency: {validation_results['logical_errors']} errors")
print(f"âœ“ Genre alignment: {validation_results['genre_alignment']:.1%}")
print(f"âœ“ Age appropriateness: {validation_results['age_appropriateness_issues']} issues")
print(f"âœ“ Movie reuse rate: {validation_results['reuse_rate']:.1f}%")

# Determine readiness for next steps
if (validation_results['logical_errors'] == 0 and 
    validation_results['genre_alignment'] >= 0.6 and 
    validation_results['age_appropriateness_issues'] <= 2 and
    validation_results['reuse_rate'] <= 35):
    print("\nðŸŽ‰ SEEDING QUALITY: READY FOR NEXT STEPS")
    print("The improved seeding meets quality thresholds for ML training")
else:
    print("\nâš ï¸  SEEDING QUALITY: NEEDS FURTHER REFINEMENT")
    print("Consider additional fine-tuning before proceeding to SMOTE/ML training")

print("\n=== COMPREHENSIVE VALIDATION COMPLETE ===")
```

```python
def seed_survey_FOOLPROOF(survey_df, finished_movies, unfinished_movies):
    """
    FOOLPROOF movie seeding - manually assign different movies to each respondent
    NO randomness, NO complex logic, just different movies for each person
    """
    print("=== FOOLPROOF MOVIE SEEDING (MANUAL ASSIGNMENT) ===")
    
    # Create a copy of the survey dataframe
    seeded_df = survey_df.copy()
    
    # Initialize the new columns
    finished_cols = [f'finished_{i}' for i in range(1, 6)]
    unfinished_cols = [f'unfinished_{i}' for i in range(1, 6)]
    
    for col in finished_cols + unfinished_cols:
        seeded_df[col] = None
    
    # Get the dropoff column
    dropoff_column = 'Have you ever started watching a movie but did not finish it?'
    
    # Convert to lists for easy indexing - use correct column name
    finished_titles = finished_movies['movie_title'].tolist()
    unfinished_titles = unfinished_movies['movie_title'].tolist()
    
    print(f"Available finished movies: {len(finished_titles)}")
    print(f"Available unfinished movies: {len(unfinished_titles)}")
    
    # Process each respondent with MANUAL assignment
    for idx, respondent in seeded_df.iterrows():
        print(f"Processing respondent {idx + 1}/{len(seeded_df)}")
        
        # Calculate starting index for this respondent to ensure different movies
        start_idx_finished = (idx * 5) % len(finished_titles)
        start_idx_unfinished = (idx * 5) % len(unfinished_titles)
        
        # Assign finished movies (5 consecutive movies starting from calculated index)
        for i in range(5):
            movie_idx = (start_idx_finished + i) % len(finished_titles)
            movie_title = finished_titles[movie_idx]
            seeded_df.at[idx, f'finished_{i+1}'] = movie_title
        
        # Assign unfinished movies only if respondent answered "Yes"
        if respondent[dropoff_column] == 'Yes':
            for i in range(5):
                movie_idx = (start_idx_unfinished + i) % len(unfinished_titles)
                movie_title = unfinished_titles[movie_idx]
                seeded_df.at[idx, f'unfinished_{i+1}'] = movie_title
    
    return seeded_df

# Execute FOOLPROOF seeding
print("Starting FOOLPROOF movie seeding...")
seeded_survey = seed_survey_FOOLPROOF(survey_df, finished_movies, unfinished_movies)

print("\n=== FOOLPROOF SEEDING COMPLETE ===")
print(f"Original survey shape: {survey_df.shape}")
print(f"Seeded survey shape: {seeded_survey.shape}")

# IMMEDIATE verification
print("\n=== IMMEDIATE VERIFICATION ===")
print("First 3 respondents' finished movies:")
for i in range(min(3, len(seeded_survey))):
    movies = []
    for j in range(1, 6):
        movie = seeded_survey.iloc[i][f'finished_{j}']
        if pd.notna(movie):
            movies.append(movie)
    print(f"  Respondent {i+1}: {movies}")

# Check if they're different
resp1_movies = [seeded_survey.iloc[0][f'finished_{j}'] for j in range(1, 6)]
resp2_movies = [seeded_survey.iloc[1][f'finished_{j}'] for j in range(1, 6)]

if set(resp1_movies) == set(resp2_movies):
    print("âŒ STILL FAILED - IDENTICAL MOVIES!")
else:
    print("âœ… SUCCESS - DIFFERENT MOVIES!")

# Save the FOOLPROOF version
seeded_survey.to_csv('../seeded_survey_data_FIXED.csv', index=False)
print("âœ… Saved as '../seeded_survey_data_FIXED.csv'")
```

```python
def analyze_movie_assignments(seeded_df, movies_df):
    """
    Analyze the characteristics of assigned movies for quality assurance
    """
    print("=== QUALITY ASSURANCE ANALYSIS ===")
    
    # Collect all assigned movies
    finished_movies_assigned = []
    unfinished_movies_assigned = []
    
    finished_cols = [f'finished_{i}' for i in range(1, 6)]
    unfinished_cols = [f'unfinished_{i}' for i in range(1, 6)]
    
    # Gather all assigned movies
    for col in finished_cols:
        finished_movies_assigned.extend(seeded_df[col].dropna().tolist())
    
    for col in unfinished_cols:
        unfinished_movies_assigned.extend(seeded_df[col].dropna().tolist())
    
    # Get movie details for analysis
    finished_details = movies_df[movies_df['movie_title'].isin(finished_movies_assigned)]
    unfinished_details = movies_df[movies_df['movie_title'].isin(unfinished_movies_assigned)]
    
    print(f"Total finished movies assigned: {len(finished_movies_assigned)}")
    print(f"Unique finished movies: {len(set(finished_movies_assigned))}")
    print(f"Total unfinished movies assigned: {len(unfinished_movies_assigned)}")
    print(f"Unique unfinished movies: {len(set(unfinished_movies_assigned))}")
    
    # Analyze ratings - use correct column name
    print("\n=== RATING ANALYSIS ===")
    print("Finished movies rating stats:")
    print(finished_details['imdb_score'].describe())
    
    if len(unfinished_details) > 0:
        print("\nUnfinished movies rating stats:")
        print(unfinished_details['imdb_score'].describe())
    
    # Analyze runtime - use correct column name
    print("\n=== RUNTIME ANALYSIS ===")
    print("Finished movies runtime stats:")
    print(finished_details['duration'].describe())
    
    if len(unfinished_details) > 0:
        print("\nUnfinished movies runtime stats:")
        print(unfinished_details['duration'].describe())
    
    # Analyze genres
    print("\n=== GENRE ANALYSIS ===")
    
    def get_genre_distribution(movie_details):
        all_genres = []
        for genre_list in movie_details['genres'].dropna():
            all_genres.extend([g.strip() for g in genre_list.split(',')])
        return pd.Series(all_genres).value_counts()
    
    finished_genres = get_genre_distribution(finished_details)
    print("Top genres in finished movies:")
    print(finished_genres.head(10))
    
    if len(unfinished_details) > 0:
        unfinished_genres = get_genre_distribution(unfinished_details)
        print("\nTop genres in unfinished movies:")
        print(unfinished_genres.head(10))
    
    # Check for duplicates within respondents
    print("\n=== DUPLICATE CHECK ===")
    duplicate_count = 0
    for idx, row in seeded_df.iterrows():
        all_movies = []
        for col in finished_cols + unfinished_cols:
            if pd.notna(row[col]):
                all_movies.append(row[col])
        
        if len(all_movies) != len(set(all_movies)):
            duplicate_count += 1
    
    print(f"Respondents with duplicate movies: {duplicate_count}")
    
    return finished_details, unfinished_details

# Run the analysis
finished_analysis, unfinished_analysis = analyze_movie_assignments(seeded_survey, movies_df)
```
